Pycity school analysis
Based on the data gathered in the size_summary cell, Smaller school tend to have a higher average score in both reading and math compared to larger schools. 

Based on the data gathered in the spending_summary cell, As school budget decreases, average peformance in both reading a math improved compared to schools with larger budgets. This is suggesting that higher spending does not correlate with academic performance. Academic performance is likely associated with other factors outside of school resources.

```python
# Standard init
import pandas as pd
from pathlib import Path 

# Files to load
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Reading the data and storing it
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Merge the data
school_data_complete = pd.merge(student_data, school_data, how="left", on="school_name")
school_data_complete.head()
                         
```

```python
# Total number of unique schools
total_schools = school_data_complete['school_name'].nunique()

# Total number of students
total_students = school_data_complete['Student ID'].nunique()

# Total budget (assuming "budget" is per school in the school_data)
total_budget = school_data_complete.groupby('school_name')['budget'].first().sum()

# Average math and reading scores
average_math_score = school_data_complete['math_score'].mean()
average_reading_score = school_data_complete['reading_score'].mean()

# Percentage passing math (assuming passing is a score >= 70)
passing_math = school_data_complete[school_data_complete['math_score'] >= 70].count()['Student ID']
percent_passing_math = (passing_math / total_students) * 100

# Percentage passing reading (assuming passing is a score >= 70)
passing_reading = school_data_complete[school_data_complete['reading_score'] >= 70].count()['Student ID']
percent_passing_reading = (passing_reading / total_students) * 100

# Percentage passing both math and reading
passing_both = school_data_complete[
    (school_data_complete['math_score'] >= 70) & (school_data_complete['reading_score'] >= 70)
].count()['Student ID']
percent_overall_passing = (passing_both / total_students) * 100

# Create a high-level summary DataFrame
district_summary = pd.DataFrame({
    "Total Schools": [total_schools],
    "Total Students": [total_students],
    "Total Budget": [total_budget],
    "Average Math Score": [average_math_score],
    "Average Reading Score": [average_reading_score],
    "% Passing Math": [percent_passing_math],
    "% Passing Reading": [percent_passing_reading],
    "% Overall Passing": [percent_overall_passing]
})

# Display the summary DataFrame
district_summary
```

```python
# Group data by school to aggregate metrics
school_group = school_data_complete.groupby(['school_name', 'type'])

# Total students per school
total_students_per_school = school_group.size()

# Total budget per school (assuming each row has the total school budget, we use `.first()` to avoid duplication)
total_budget_per_school = school_group['budget'].first()

# Per student budget
per_student_budget = total_budget_per_school / total_students_per_school

# Average math score per school
average_math_score_per_school = school_group['math_score'].mean()

# Average reading score per school
average_reading_score_per_school = school_group['reading_score'].mean()

# % Passing Math (students with math score >= 70)
passing_math_per_school = school_data_complete[school_data_complete['math_score'] >= 70].groupby('school_name').size()
percent_passing_math = (passing_math_per_school / total_students_per_school) * 100

# % Passing Reading (students with reading score >= 70)
passing_reading_per_school = school_data_complete[school_data_complete['reading_score'] >= 70].groupby('school_name').size()
percent_passing_reading = (passing_reading_per_school / total_students_per_school) * 100

# % Overall Passing (students passing both math and reading)
passing_both_per_school = school_data_complete[
    (school_data_complete['math_score'] >= 70) & (school_data_complete['reading_score'] >= 70)
].groupby('school_name').size()
percent_overall_passing = (passing_both_per_school / total_students_per_school) * 100

# Create a summary DataFrame
school_summary = pd.DataFrame({
    "School Type": school_group['type'].first(),
    "Total Students": total_students_per_school,
    "Total School Budget": total_budget_per_school,
    "Per Student Budget": per_student_budget,
    "Average Math Score": average_math_score_per_school,
    "Average Reading Score": average_reading_score_per_school,
    "% Passing Math": percent_passing_math,
    "% Passing Reading": percent_passing_reading,
    "% Overall Passing": percent_overall_passing
})

# Display the summary DataFrame
school_summary.reset_index(inplace=True)  # To flatten the index
school_summary
```

```python
# Sort the schools by '% Overall Passing' in descending order
top_schools = school_summary.sort_values(by='% Overall Passing', ascending=False)

# Display the top 5 rows
top_schools.head()
```

```python
# Sort the schools by '% Overall Passing' in ascending order
bottom_schools = school_summary.sort_values(by='% Overall Passing', ascending=True)

# Display the top 5 rows (the bottom 5 schools based on overall passing rate)
bottom_schools.head()
```

```python
# Grouping the data by 'school name' and 'grade' to calculate average math score per grade per school
average_math_score_by_grade = school_data_complete.groupby(['school_name', 'grade'])['math_score'].mean()

# Convert the Series to a DataFrame
average_math_score_by_grade = average_math_score_by_grade.reset_index()

# Renaming the column for clarity
average_math_score_by_grade = average_math_score_by_grade.rename(columns={'math_score': 'Average Math Score'})

# Display the DataFrame
average_math_score_by_grade
```

```python
# Grouping the data by 'school name' and 'grade' to calculate the average reading score per grade per school
average_reading_score_by_grade = school_data_complete.groupby(['school_name', 'grade'])['reading_score'].mean()

# Convert the Series to a DataFrame
average_reading_score_by_grade = average_reading_score_by_grade.reset_index()

# Renaming the column for clarity
average_reading_score_by_grade = average_reading_score_by_grade.rename(columns={'reading_score': 'Average Reading Score'})

# Display the DataFrame
average_reading_score_by_grade
```

```python
# Define spending bins and labels
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]

# Create a new column in the DataFrame for spending ranges based on 'Per Student Budget'
school_summary['Spending Ranges (Per Student)'] = pd.cut(school_summary['Per Student Budget'], bins=spending_bins, labels=labels)

# Group the data by the 'Spending Ranges (Per Student)' and calculate the mean for each metric
spending_summary = school_summary.groupby('Spending Ranges (Per Student)', observed=False).agg({
    'Average Math Score': 'mean',
    'Average Reading Score': 'mean',
    '% Passing Math': 'mean',
    '% Passing Reading': 'mean',
    '% Overall Passing': 'mean'
})

# Display the summary DataFrame
spending_summary
```

```python
# Define spending bins and labels
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]

# Use pd.cut to categorize spending based on the bins
school_summary['Spending Ranges (Per Student)'] = pd.cut(school_summary['Per Student Budget'], bins=spending_bins, labels=labels)

# Group by 'Spending Ranges (Per Student)' and calculate mean scores with observed=False to retain current behavior
spending_math_scores = school_summary.groupby("Spending Ranges (Per Student)", observed=False)["Average Math Score"].mean()
spending_reading_scores = school_summary.groupby("Spending Ranges (Per Student)", observed=False)["Average Reading Score"].mean()
spending_passing_math = school_summary.groupby("Spending Ranges (Per Student)", observed=False)["% Passing Math"].mean()
spending_passing_reading = school_summary.groupby("Spending Ranges (Per Student)", observed=False)["% Passing Reading"].mean()
overall_passing_spending = school_summary.groupby("Spending Ranges (Per Student)", observed=False)["% Overall Passing"].mean()

# Combine all these metrics into a single DataFrame for easy visualization
spending_summary = pd.DataFrame({
    "Average Math Score": spending_math_scores,
    "Average Reading Score": spending_reading_scores,
    "% Passing Math": spending_passing_math,
    "% Passing Reading": spending_passing_reading,
    "% Overall Passing": overall_passing_spending
})

# Display the summary DataFrame
spending_summary
```

```python
# Define the size bins and labels
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

# Use pd.cut to categorize school size based on the bins
school_summary['School Size'] = pd.cut(school_summary['Total Students'], bins=size_bins, labels=labels)

# Group by 'School Size' and calculate mean scores and passing percentages
size_math_scores = school_summary.groupby("School Size", observed=False)["Average Math Score"].mean()
size_reading_scores = school_summary.groupby("School Size", observed=False)["Average Reading Score"].mean()
size_passing_math = school_summary.groupby("School Size", observed=False)["% Passing Math"].mean()
size_passing_reading = school_summary.groupby("School Size", observed=False)["% Passing Reading"].mean()
size_overall_passing = school_summary.groupby("School Size", observed=False)["% Overall Passing"].mean()

# Combine the calculated metrics into a new DataFrame called 'size_summary'
size_summary = pd.DataFrame({
    "Average Math Score": size_math_scores,
    "Average Reading Score": size_reading_scores,
    "% Passing Math": size_passing_math,
    "% Passing Reading": size_passing_reading,
    "% Overall Passing": size_overall_passing
})

# Display the 'size_summary' DataFrame
size_summary
```
